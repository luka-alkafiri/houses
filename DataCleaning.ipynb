**Populate Property Address data**

```sql
UPDATE a
SET PropertyAddress = ISNULL(a.PropertyAddress,b.PropertyAddress)
FROM portfolio.dbo.houses a
JOIN portfolio.dbo.houses b
    ON a.ParcelID = b.ParcelID
    AND a.UniqueID <> b.UniqueID
WHERE a.PropertyAddress IS NULL
```

**Breaking out Address into Individual Columns (Address, City, State)**

```sql
ALTER TABLE portfolio.dbo.houses
ADD City nvarchar(50), Address nvarchar(50), State nvarchar(50)
```

```sql
UPDATE portfolio.dbo.houses
SET Address = SUBSTRING(PropertyAddress, 1, CHARINDEX(',', PropertyAddress) -1)
```

```sql
UPDATE portfolio.dbo.houses
SET City = SUBSTRING(PropertyAddress, CHARINDEX(',', PropertyAddress) +1, LEN(PropertyAddress))
```

```sql
UPDATE portfolio.dbo.houses
SET State = PARSENAME(REPLACE(OwnerAddress, ',', '.') , 1)
```

**Change Y and N to Yes and No in "Sold as Vacant" field**

```sql
UPDATE portfolio.dbo.houses
SET SoldAsVacant = CASE When SoldAsVacant = 'Y' THEN 'Yes'
	   When SoldAsVacant = 'N' THEN 'No'
	   ELSE SoldAsVacant
	   END
```

```sql
SELECT SoldAsVacant
FROM portfolio.dbo.houses
WHERE SoldAsVacant = 'N' OR SoldAsVacant = 'Y'
```

**Remove Duplicates**

```sql
SELECT TOP 10
    *,
	ROW_NUMBER() OVER (
	PARTITION BY ParcelID,
				 PropertyAddress,
				 SalePrice,
				 SaleDate,
				 LegalReference
				 ORDER BY
					UniqueID
					) row_num

FROM portfolio.dbo.houses
ORDER BY row_num DESC
```

```sql
WITH UniqueCTE AS (
    SELECT 
    *,
	ROW_NUMBER() OVER (
	PARTITION BY ParcelID,
				 PropertyAddress,
				 SalePrice,
				 SaleDate,
				 LegalReference
				 ORDER BY
					UniqueID
					) row_num

FROM portfolio.dbo.houses
)

SELECT *
FROM UniqueCTE
WHERE row_num > 1
ORDER BY PropertyAddress
```

```sql
WITH UniqueCTE AS (
    SELECT 
    *,
	ROW_NUMBER() OVER (
	PARTITION BY ParcelID,
				 PropertyAddress,
				 SalePrice,
				 SaleDate,
				 LegalReference
				 ORDER BY
					UniqueID
					) row_num

FROM portfolio.dbo.houses
)

DELETE
FROM UniqueCTE
WHERE row_num > 1
```

```sql
WITH UniqueCTE AS (
    SELECT 
    *,
	ROW_NUMBER() OVER (
	PARTITION BY ParcelID,
				 PropertyAddress,
				 SalePrice,
				 SaleDate,
				 LegalReference
				 ORDER BY
					UniqueID
					) row_num

FROM portfolio.dbo.houses
)

SELECT *
FROM UniqueCTE
WHERE row_num > 1
ORDER BY PropertyAddress
```
